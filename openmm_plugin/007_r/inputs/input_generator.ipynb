```python
import numpy as np
import mdtraj as mdj
from BFEE2.commonTools import fileParser
```

```python
selectionRef = "protein"
selectionLig = "resname MOL"
topFile = 'complex_largebox.prmtop'
coorFile = 'complex_largebox_npt.rst7'
fParser = fileParser.fileParser(topFile, coorFile)
if (fParser.measurePolarAngles(selectionRef, selectionLig)[0] > 150 or fParser.measurePolarAngles(selectionRef, selectionLig)[0] < 30):
        fParser.rotateSystem('x', 90)
    # write the new topology and structure file
fParser.saveFile('all', f'complex_largebox_bfee2.pdb', 'pdb', False)
```

```python
pdb = mdj.load_pdb('complex_largebox_bfee2.pdb')
# protein and type!="H"'
protein = pdb.topology.select('protein and type!="H"')
bfactors = np.ones(protein.size)
protein = pdb.atom_slice(protein)
serials = [atom.serial for atom in protein.topology.atoms]
print(str(serials).replace(' ', ''))
protein.save_pdb('protein.pdb', bfactors=bfactors)
```

```python
pdb = mdj.load_pdb('complex_largebox_bfee2.pdb')
ligand = pdb.topology.select('resname "MOL" and type!="H"')
bfactors = np.ones(ligand.size)
ligand = pdb.atom_slice(ligand)
serials = [atom.serial for atom in ligand.topology.atoms]
print(str(serials).replace(' ', ''))
ligand.save_pdb('ligand.pdb', bfactors=bfactors)
```

```python
def measure_center(atom_positions):
    """mimic the VMD command "measure center"
    Args:
        atom_positions (numpy.array): a numpy array containing the XYZ coordinates of N atoms. The shape should be (N, 3).
    Returns:
        Numpy.array: a shape of (3,) array contains the geometric center
    """    

    xyz_array = np.transpose(atom_positions)
    center_x = np.average(xyz_array[0])
    center_y = np.average(xyz_array[1])
    center_z = np.average(xyz_array[2])
    return np.array([center_x, center_y, center_z])
```

```python
pdb_crys = mdj.load('complex_largebox_bfee2.pdb')
protein_idxs = pdb_crys.top.select("protein and type!='H'")
pdb_crys_com = measure_center(pdb_crys.xyz[:, protein_idxs, :])
# 
print(str(list(pdb_crys_com)).replace(' ', ''))
```
